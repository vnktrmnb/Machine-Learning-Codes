```python
def solve_cramers(A, B):
    # Compute the determinant of the coefficient matrix A
    det_A = np.linalg.det(A)

    # Check if the determinant is zero (which indicates the matrix is singular)
    if det_A == 0:
        print("Determinant of coefficient matrix is zero. Cannot use Cramer's rule.")
        return None

    # Get the number of unknowns (size of the system)
    n = len(A)

    # Initialize an empty list to store the solutions
    solutions = []

    # Iterate over each unknown (column of the coefficient matrix)
    for i in range(n):
        # Create a copy of the coefficient matrix and replace the i-th column with the constant vector B
        A_copy = A.copy()
        A_copy[:, i] = B

        # Compute the determinant of the modified matrix
        det_Ai = np.linalg.det(A_copy)

        # Compute the solution for the i-th unknown using Cramer's rule
        x_i = det_Ai / det_A

        # Append the solution to the list of solutions
        solutions.append(x_i)

    # Return the list of solutions
    return solutions
```

```python
# Solve the system of equations using Cramer's rule
solutions = solve_cramers(A, B)

```
