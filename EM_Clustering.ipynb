```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

# Given set of 8 objects with (x, y) coordinates
data = np.array([
    [1, 2],
    [2, 3],
    [3, 3],
    [6, 8],
    [7, 9],
    [8, 8],
    [9, 7],
    [10, 9]
])

# Plot the dataset
plt.scatter(data[:, 0], data[:, 1], s=50, cmap='viridis')
plt.title('Scatter Plot of the Data Points')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

# Fitting the Gaussian Mixture Model with 2 clusters
gmm = GaussianMixture(n_components=2, random_state=0)
gmm.fit(data)

# Predict the cluster for each data point
labels = gmm.predict(data)

# Plot the clustered data
plt.scatter(data[:, 0], data[:, 1], c=labels, s=50, cmap='viridis')

# Plot the centers of the clusters
centers = gmm.means_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title('Clustered Data with GMM')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
