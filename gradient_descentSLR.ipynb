```python

import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(x,y,iterations,learning_rate):

    B0_curr = 0.0
    B1_curr = 0.0

    n = len(x)

    for i in range(iterations):

        y_predicted = B0_curr + B1_curr * x

        B0d = -(2/n)*sum(y-y_predicted)
        B1d = -(2/n)*sum(x*(y-y_predicted))

        loss = (1/n) * sum([val**2 for val in (y-y_predicted)])

        B0_curr = B0_curr - learning_rate * B0d
        B1_curr = B1_curr - learning_rate * B1d

        plotline(x,y,y_predicted)

        print ("iteration {}, B0 {}, B1 {}, loss {}".format(i,B0_curr,B1_curr,loss))

```

```python
def plotline(x,y,y_predicted):
  ax = plt.axes()
  ax.scatter(x,y,color='red')
  ax.plot(x,y_predicted,color='green')
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  plt.show()



```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,3,4,5,6])

iterations = 100
learning_rate = 0.01

gradient_descent(x,y,iterations,learning_rate)
```
